Let's pull the data for constituents and drop some columns to work with a more solid data frame.

```python
const = pd.read_html("https://en.wikipedia.org/wiki/Dow_Jones_Industrial_Average")[1]
```

```python
const = const.iloc[:, :5].copy()
const
```

White space in the column name is not a good idea. Let's add underscore. Lastlt, for the data columns, let's change the type to datatime.

```python
const.rename(columns={"Date added": "Date_Added"},inplace=True)
```

```python
const.Date_Added = pd.to_datetime(const.Date_Added)
```

```python
const.info()
```

### Let's Normalize Unicode Strings and Get the Ticker Symbols

The problem stemmed from the data coming originally from a html file. We should encode and normalize unicode strings into a normal form.

```python
import unicodedata
```

```python
const.Symbol
```

```python
const.Symbol.to_list()
```

```python
const.Symbol.apply(lambda x: unicodedata.normalize("NFKD", x))
```

```python
const.Symbol = const.Symbol.apply(lambda x: unicodedata.normalize("NFKD", x))
```

```python
const["Ticker"] = const.Symbol.str.split(": ").apply(lambda x: x[-1])
```

```python
const
```

```python
ticker_list = const.Ticker.to_list()
```

```python
ticker_list
```

```python
const.to_csv("const.csv", index = False)
```

### Let's Load and Save Historical Stock Prices

Let's import proper libraries. We will pull data from Yahoo so we need to yfinance package.

```python
import pandas as pd
import yfinance as yf
```

```python
ticker_list
```

```python
prices = yf.download(ticker_list, start = "2007-01-01", end = "2020-03-31")
```

```python
prices
```

```python
prices = prices.loc[:,"Close"].copy()
```

```python
prices
```

```python
prices.to_csv("const_prices.csv")
```
